```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('./science.mplstyle')
```

```python
# Load the log file
data = pd.read_csv('../data/model/training_log.csv')

# Extract relevant metrics
epochs = data['epoch']
train_loss = data['train_loss']
val_loss = data['val_loss']

# Plot training and validation loss
plt.figure(figsize=(10, 8))
plt.scatter(epochs, train_loss, label='Training Loss', color='blue', linewidth=2)
plt.scatter(epochs, val_loss, label='Validation Loss', color='orange', linewidth=2)

# Add labels, legend, and grid
#plt.title("Model Training Progress (Steps)", fontsize=16)
plt.xlabel("Epoch", fontsize=20)
plt.ylabel("Loss", fontsize=20)
plt.legend(fontsize=20)
#plt.ylim(1e-4, 1e-2)
plt.yscale('log')

```
